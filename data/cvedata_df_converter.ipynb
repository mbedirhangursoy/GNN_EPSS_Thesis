```python
cve_data = [
    {"cve_id": "CVE-2025-2095", "cvss": 6.3, "cwe": "CWE-78", "vendor": "MICROSOFT", "EPSS_score": 0.1, "date_published": "2025-03-07", "severity": "MEDIUM", 'sentence': "A vulnerability classified as critical has been found in TOTOLINK EX1800T 9.1.0cu.2112_B20220316. This affects the function setDmzCfg of the file /cgi-bin/cstecgi.cgi. The manipulation of the argument ip leads to os command injection. It is possible to initiate the attack remotely. The exploit has been disclosed to the public and may be used."},
    {"cve_id": "CVE-2025-3001", "cvss": 6.5, "cwe": "CWE-77", "vendor": "MICROSOFT", "EPSS_score": 0.3, "date_published": "2025-03-10", "severity": "MEDIUM", 'sentence': "A severe vulnerability has been identified in the TOTOLINK EX1800T 9.1.0cu.2112_B20220316, impacting the setDmzCfg function in the /cgi-bin/cstecgi.cgi file."},
    {"cve_id": "CVE-2025-4012", "cvss": 7.0, "cwe": "CWE-78", "vendor": "NETGEAR", "EPSS_score": 0.5, "date_published": "2025-03-15", "severity": "HIGH", 'sentence': 'A critical security flaw has been discovered in the TOTOLINB EX1800F router, which allows remote attackers to inject OS commands through the manipulation of specific parameters in the deviceâ€™s configuration file.'},

]

def df_converter(data): #testing only on 2025 0xxx
    with open(data) as file:
        df = pd.read_json(file)
        id = df['cveMetadata']['cveId']
        try:
            metrics = df['containers']['cna']['metrics'][0]['cvssV3_1']
        except KeyError:
            try:
                metrics = df['containers']['cna']['metrics'][0]['cvssV3_0']
            except KeyError:
                try:
                    metrics = df['containers']['adp'][0]['metrics'][0]['cvssV3_1']
                except KeyError:    
                    metrics = {}


        baseseverity = metrics.get('baseSeverity')
        basescore = metrics.get('baseScore')
        confidentialityimpact = metrics.get('confidentialityImpact')
        integrityimpact = metrics.get('integrityImpact')


        try:
            cwe =  df['containers']['cna']['problemTypes'][0]['descriptions'][0]['cweId']
        except KeyError:
            try:
                cwe =  df['containers']['adp'][0]['problemTypes'][0]['descriptions'][0]['cweId']
            except KeyError:
                cwe = None

        try:
            vendor =  df['containers']['cna']['affected'][0]['vendor']
            if vendor == 'n/a':
                vendor = None
        except KeyError:
            vendor = None

        try:
            description =  df['containers']['cna']['descriptions'][0]['value']
        except KeyError:
            description = None

       
        return id, basescore, baseseverity, confidentialityimpact, integrityimpact, vendor, description, cwe


#df_converter('cvelistV5-main/cves/2025/28xxx/CVE-2025-28144.json')
```

```python
def create_df_all_data(start_year, end_year):
    cve_data = []
    

    while start_year <= end_year:
        for filesxx in os.listdir(f'cvelistV5-main/cves/{start_year}/'):
            ds_store_file_location = f'cvelistV5-main/cves/{start_year}/.DS_store' #remove all hidden files
            if os.path.isfile(ds_store_file_location):
                os.remove(ds_store_file_location)
            for cve in os.listdir(f'cvelistV5-main/cves/{start_year}/{filesxx}'):
                print(cve)
                id, basescore, baseseverity, confidentialityimpact, integrityimpact, vendor, description, cwe = df_converter(f'cvelistV5-main/cves/{start_year}/{filesxx}/{cve}')
                temp_dict = {}
                temp_dict['cve_id'] = id
                temp_dict['basescore'] = basescore
                temp_dict['baseseverity'] = baseseverity
                temp_dict['confidentialityimpact'] = confidentialityimpact
                temp_dict['integrity'] = integrityimpact
                temp_dict['vendor'] = vendor
                temp_dict['description'] = description
                temp_dict['CWE'] = cwe
                cve_data.append(temp_dict)

            
        start_year += 1

    df = pd.DataFrame(cve_data)
    df.to_csv('output.csv')
    return df
```
